# Json to csv

```python
import os
import pandas as pd
import numpy
import json
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import MinMaxScaler

from pandas import json_normalize

from numpy.random import randn
from numpy.random import seed
from scipy.stats import spearmanr

t_list = []
ut_list = []


t=0
ut=0
count = 1
rname = []
data_list = []



cname = ["rating", "sentiment", "review","rating std","sentiment std","mae","spearman","sympathy","unsympathy","count"]
file_list = os.listdir('C:/Users/User/Desktop/four movies/outlaws')
num1 = len(file_list)

for i in range(0, num1):     
    file = "C:/Users/User/Desktop/four movies/outlaws/%d.json" % i
    try:
        with open(file, encoding="UTF-8") as json_file:
            json_data = json.load(json_file)
            num = len(json_data["data"])
            if(num>=3 and (json_data["sympathy"]!=json_data["unsympathy"])):
                dt = []
                ct = 0
                t_1 = [] # rating, predict
                t_2 = [] # sentiment score, test
                t_3 = []

                for x in range(0,num):
                    if(json_data["data"][x]["title"]=="범죄도시"):
                        if(ct == 0):
                            if(len(json_data["data"][x]["review"])<=140):
                                dt.append(int(json_data["data"][x]["rating"]))
                                dt.append(json_data["data"][x]["sentiment score"])
                                dt.append(json_data["data"][x]["review"])
                                ct += 1
                    t_1.append(int(json_data["data"][x]["rating"]))
                    t_2.append(json_data["data"][x]["sentiment score"])
                    t_3.append(json_data["data"][x]["sentiment score"]*10)
                std1 = numpy.std(t_1)
                std2 = numpy.std(t_2)
                #df = json_normalize(json_data["data"])
                #df['rating'] = pd.to_numeric(df['rating'])
                #t_1 = df["rating"].to_numpy()
                #t_2 = df["sentiment score"].to_numpy()
                #transformer = MinMaxScaler(feature_range=(0, 10))
                #t_2 = transformer.fit_transform(t_2)
                spearman, _ = spearmanr(t_1, t_3)
                mae = mean_absolute_error(t_1, t_3)
                dt.append(std1)
                dt.append(std2)
                dt.append(mae)
                dt.append(spearman)
                dt.append(json_data["sympathy"])
                dt.append(json_data["unsympathy"])
                dt.append(count)
                rname.append(count)
                data_list.append(dt)
                count += 1
                
                    
    except Exception:
            print("%i.json은 존재하지 않는 파일입니다." % i)
            
pdf2 = pd.DataFrame(data_list, columns=cname)
mname = "./result.csv"
pdf2 = pdf2.dropna(axis=0)
pdf2.to_csv(mname,encoding="utf-8-sig")

import os
import pandas as pd
import numpy
import json
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import MinMaxScaler

from pandas import json_normalize

from numpy.random import randn
from numpy.random import seed
from scipy.stats import spearmanr

t_list = []
ut_list = []


t=0
ut=0
count = 1
rname = []
data_list = []



cname = ["rating", "sentiment", "review","rating std","sentiment std","mae","spearman","sympathy","unsympathy","count"]
file_list = os.listdir('C:/Users/User/Desktop/four movies/taxi')
num1 = len(file_list)

for i in range(0, num1):     
    file = "C:/Users/User/Desktop/four movies/taxi/%d.json" % i
    try:
        with open(file, encoding="UTF-8") as json_file:
            json_data = json.load(json_file)
            num = len(json_data["data"])
            if(num>=3 and (json_data["sympathy"]!=json_data["unsympathy"])):
                dt = []
                ct = 0
                t_1 = [] # rating, predict
                t_2 = [] # sentiment score, test
                t_3 = []

                for x in range(0,num):
                    if(json_data["data"][x]["title"]=="택시운전사"):
                        if(ct == 0):
                            if(len(json_data["data"][x]["review"])<=140):
                                dt.append(int(json_data["data"][x]["rating"]))
                                dt.append(json_data["data"][x]["sentiment score"])
                                dt.append(json_data["data"][x]["review"])
                                ct += 1
                    t_1.append(int(json_data["data"][x]["rating"]))
                    t_2.append(json_data["data"][x]["sentiment score"])
                    t_3.append(json_data["data"][x]["sentiment score"]*10)
                std1 = numpy.std(t_1)
                std2 = numpy.std(t_2)
                #df = json_normalize(json_data["data"])
                #df['rating'] = pd.to_numeric(df['rating'])
                #t_1 = df["rating"].to_numpy()
                #t_2 = df["sentiment score"].to_numpy()
                #transformer = MinMaxScaler(feature_range=(0, 10))
                #t_2 = transformer.fit_transform(t_2)
                spearman, _ = spearmanr(t_1, t_3)
                mae = mean_absolute_error(t_1, t_3)
                dt.append(std1)
                dt.append(std2)
                dt.append(mae)
                dt.append(spearman)
                dt.append(json_data["sympathy"])
                dt.append(json_data["unsympathy"])
                dt.append(count)
                rname.append(count)
                data_list.append(dt)
                count += 1
                
                    
    except Exception:
            print("%i.json은 존재하지 않는 파일입니다." % i)
            
pdf2 = pd.DataFrame(data_list, columns=cname, index=rname)
mname = "./definition/taxi.csv"
pdf2 = pdf2.dropna(axis=0)
pdf2.to_csv(mname,encoding="utf-8-sig")
```
